```python
#........FINDINGS.......
# It seems that value of  PL, PW is very low for iris setosa species. SO, we can conclude they have small petals compare to other two species
# For iris- virginica, pw ,pl values are high. It means they have the bigger sepals compare to other two species
#For iris-versicolor, values of PL,PW,SL,SW are in medium range...so they may have medium size petals and sepals.

```
